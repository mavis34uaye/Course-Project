```python
mplt.hist(df['quality'],bins=6, color='blue', alpha=0.7)
mplt.xlabel('Wine Quality')
mplt.ylabel('Frequency')
mplt.title('Histogram of Wine Quality Scores')
mplt.show()
```

```python
sns.pairplot(df, vars=['alcohol', 'pH', 'residual sugar', 'quality'], hue='quality')
mplt.suptitle('Pair Plot of Wine Attributes')
mplt.show()
```

```python
quality_distribution = df['quality'].value_counts()
mplt.figure(figsize=(8, 8))
mplt.pie(quality_distribution, labels=quality_distribution.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
mplt.title('Distribution of Wine Quality')
mplt.show()
```
